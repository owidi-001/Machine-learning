```python
from matplotlib import pyplot as plt
#
import pandas as pd
from sklearn.neighbors.classification import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

data=pd.read_csv('winequalityN.csv')
```

```python
data.head(10)
```

```python
data.tail(10)
```

```python
# logical regression....using binary representation
def process(n):
    if n>6.5:
        return 1
    else:
        return 0
```

```python
data["quality"]=data.quality.apply(process)
data['quality']
```

```python
data=data.dropna()
data.tail(100)
```

```python
# transforming type to numerical values
def wine_flavour(text):
    if text=='red':
        return 1
    else:
        return 0
```

```python
data.type=data.type.apply(wine_flavour)
data.type
```

```python
# split data into x and y
y=data.quality #data['quality']
x=data.drop(['quality'],axis=1)
```

```python
# split data for testing and training
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
x_train.count()
y_train.count()
x_test.count()
y_test.count()
data.head(100)
```

```python
# scale values
se=StandardScaler()
x_train=se.fit_transform(x_train)
x_test=se.fit_transform(x_test)
x_train
```

```python
neighbors??
```

```python
# create a model
model=KNeighborsClassifier()
```

```python
# train your model
model.fit(x_train,y_train)
```

```python
# predict
prediction=model.predict(x_test)
prediction
```

```python
print(classification_report(y_test,prediction))
```
