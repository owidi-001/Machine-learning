```python
from matplotlib import pyplot as plt,style
style.use('ggplot')

x=[1,2,3]
y=[4,5,1]

x2=[2,6,3]
y2=[3,2,4]

plt.scatter(x,y,marker='o',color='k',label='dark')
plt.scatter(x2,y2,marker='x',color='g',label='green')
plt.legend()
plt.show()
```
